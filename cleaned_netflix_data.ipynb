# Drop Duplicates

```python
df.drop_duplicates()
```

# Handle Missing Values

```python
df['country'] = df['country'].fillna('Unknown')
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Not listed')
df
```

# Convert Date Columns

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year.astype('Int64')
df['month_added'] = df['date_added'].dt.month.astype('Int64')
df
```

# Clean and Separate Columns

```python
df[['duration_num', 'duration_type']] = df['duration'].str.extract(r'(\d+)\s*(\w+)')
df['duration_num'] = pd.to_numeric(df['duration_num'], errors='coerce')
df
```

# Normalize Text Data

```python
df['title'] = df['title'].str.strip().str.title()
df['country'] = df['country'].str.strip()
df['rating'] = df['rating'].str.strip()
df
```

# Data Visualization Using Matplotlib
## Count of Movies vs TV Shows

```python
df['type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Count of Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

## Content Added Over Time

```python
df['year_added'].value_counts().plot(kind='line', marker='o')
plt.title('Content Added by Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

## Top 10 Countries with Most Titles

```python
df['country'].value_counts().head(10).sort_values(ascending=True).plot(kind='barh', color='coral')
plt.title('Top 10 Countries with Most Netflix Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

## Most Common Ratings

```python
df['rating'].value_counts().plot(kind='bar', color='green')
plt.title('Distribution of Content Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```
